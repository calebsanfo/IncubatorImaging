# Test Microscope Image Quality Pre-Trained Model
Testing the pre-trained model (model.ckpt-1000042)

# Results
- Model minimum is ~2.3 on PVCAM image, should be 0
  on in-focus image. 
- Does not detect difference after ~40 um out of focus


```python
import sys
sys.path.insert(0, '../../source')
sys.path.insert(0, '../../source/miq')

import MMCorePy
import miq
import utils
from matplotlib import pyplot as plt


focus_model = miq.get_classifier()
```

```python
# Start Camera
cam = utils.start_cam()
```

```python
# Load Micromanager
# mmc = MMCorePy.CMMCore()
# mmc.loadSystemConfiguration("../../config/MMConfig_YellenLab_ubuntu.cfg")
# print(mmc.getLoadedDevices())
# z_drive = mmc.setFocusDevice('FocusDrive')
# z_loc = mmc.getPoistion()

mmc = MMCorePy.CMMCore()
# Load Scope Config from file
mmc.loadSystemConfiguration("../../config/MMConfig_YellenLab_ubuntu.cfg")
print ("Devices loaded from config file:\n", mmc.getLoadedDevices())
print ("\n Stage at: (", mmc.getXPosition(), ",", mmc.getYPosition(), ")")
xyStage = mmc.getXYStageDevice()
zStage = mmc.setFocusDevice("FocusDrive")
z_loc = mmc.getPosition()
```

# Get focus predictions:

```python
# Scope at best focus
frame = cam.get_frame(exp_time=1).reshape(cam.sensor_size[::-1])
plt.imshow(frame, cmap="gray")
prediction = focus_model.score(frame)
print (prediction)
```

```python
# Test for steps of + 2.5 um
for i in range (30):
    mmc.setPosition(z_loc + (2.5*i))
    mmc.waitForSystem()
    frame = cam.get_frame(exp_time=1).reshape(cam.sensor_size[::-1])
    prediction = focus_model.score(frame)
    print ('Prediction at +',(2.5*i), 'um: ', prediction)
```

```python
# Test for steps of - 2.5 um
for i in range (30):
    mmc.setPosition(z_loc - (2.5*i))
    mmc.waitForSystem()
    frame = cam.get_frame(exp_time=1).reshape(cam.sensor_size[::-1])
    prediction = focus_model.score(frame)
    print ('Prediction at -',(2.5*i), 'um: ', prediction)
```

```python
# Close Camera 
utils.close_cam(cam)
```
